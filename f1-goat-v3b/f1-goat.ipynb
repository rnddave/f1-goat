## importing multiple CSV files

Initially I had set out inporting one at a time like so: 


this is time consuming, there has to be a better way, and indeed there is, we could use `os` and a loop to get all of the files in one, somewhat longer, instruction:

````python
# We need to start looking at the data (the original data is in the github repo)

import os
import pandas as pd

# we find the data here:
data_source = './data'

# we need a way of addressing the individual files once they've been imported, we could use a dictionary for that
dataframes = {}

# loop through each file
for filename in os.listdir(data_source):
    if filename.endswith('csv'):
        # we now have a filename and a filepath
        file_path = os.path.join(data_source, filename)

        # we want to store the filenames (without csv) into dictionary
        df_name = filename.replace('.csv', '')
        # and now we want to reference the actual dataframe
        dataframes[df_name] = pd.read_csv(file_path)
````
